```python
df_2=df_1.copy()
```

```python
#type(''.join(df_2['Genres'].str))
```

```python
df_2['Genres']=df_2['Genres'].str.replace(',','').str.replace("Games",'').str.replace("Entertainment",'').str.replace("Strategy","").str.replace("Role Playing","Role")
```

```python
df_2["Genres"]
```

```python
df_2['Genres']=df_2['Genres'].str.split(' ').map(lambda x: ' '.join(sorted(x)))
```

```python
df_2['Genres']
```

```python
df_2['Genres']=df_2['Genres'].str.strip()
```

```python
df_2['Genres']
```

```python
n=df_2[~df_2.Genres.str.contains('Puzzle') & ~df_2.Genres.str.contains('Action')&~df_2.Genres.str.contains('Family')&~df_2.Genres.str.contains('Education')&~df_2.Genres.str.contains('Family')&~df_2.Genres.str.contains("Adventure")&~df_2.Genres.str.contains('Board')&~df_2.Genres.str.contains("Role")].index
```

```python
n
```

```python
df_2.drop(n,inplace=True)
```

```python
df_2
```

```python
df_2.loc[df_2['Genres'].str.contains('Puzzle'),'Genres'] = 'Puzzle'
df_2.loc[df_2['Genres'].str.contains('Education'),'Genres'] = 'Education'
df_2.loc[df_2['Genres'].str.contains('Role'),'Genres'] = 'Role'
df_2.loc[df_2['Genres'].str.contains('Family'),'Genres'] = 'Family'
df_2.loc[df_2['Genres'].str.contains('Action'),'Genres'] = 'Action'
df_2.loc[df_2['Genres'].str.contains('Adventure'),'Genres'] = 'Adventure'
df_2.loc[df_2['Genres'].str.contains('Board'),'Genres'] = 'Board'
df_2.loc[df_2['Genres'].str.contains('Food'),'Genres'] = 'Food'
df_2.loc[df_2['Genres'].str.contains('sports'),'Genres'] = 'Sports'
```

```python
df_2
```

```python
df_2.rename(columns={'Average User Rating':'User_Rating'},inplace=True)
```

```python
d={}
d['Action']=df_2.User_Rating[df_2['Genres']=='Action'].mean()
d['Family']=df_2.User_Rating[df_2['Genres']=='Family'].mean()
d['Adventure']=df_2.User_Rating[df_2['Genres']=='Adventure'].mean()
d['Board']=df_2.User_Rating[df_2['Genres']=='Board'].mean()
d['Role']=df_2.User_Rating[df_2['Genres']=='Role'].mean()
d['Puzzle']=df_2.User_Rating[df_2['Genres']=='Puzzle'].mean()
d['Education']=df_2.User_Rating[df_2['Genres']=='Education'].mean()
d['Sports']=df_2.User_Rating[df_2['Genres']=='Sports'].mean()
d['Food']=df_2.User_Rating[df_2['Genres']=='Food'].mean()
```

```python
m=0
k=0
#print(d)
for i in d:
   print(i,d[i])
   if(d[i]>m):
        k=i
        m=d[i]
print("the most higher user rating is ",k,"with mean ",m)
```
